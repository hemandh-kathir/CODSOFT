# **Iris Flower Classification**

**Importing the Packages**

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
```

**Load Data Set**

```python
iris_Data = pd.read_csv('/content/IRIS.csv')
```

# **Scatter Plot**

```python
iris_Data.plot(kind='scatter',x='sepal_length', y='sepal_width') ;
plt.show()
```

**Colour-Coading for Each Flower Type**

```python
from numpy.ma import size
sns.set_style("whitegrid");
sns.FacetGrid(iris_Data,hue='species', height =6).map(plt.scatter, "sepal_length", "sepal_width").add_legend();
#sns.size=4
plt.show()
```

```python
sns.pairplot(iris_Data, hue='species', height=2).add_legend()
```

```python
X = iris_Data.drop(['species'],axis = 1)
y = iris_Data['species']
```

```python
print(X)
```

```python
print(y)
```

**Analysis the Data**

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
Data_range = list(range(1,15))
val = []
for i in Data_range:
  knn = KNeighborsClassifier(n_neighbors=i).fit(X_train, y_train)
  y_predict = knn.predict(X_test)
  val.append(metrics.accuracy_score(y_test , y_predict))
for i in val:
  print(i)
```

```python
knn = KNeighborsClassifier(n_neighbors=1).fit(X_train, y_train)
y_predict = knn.predict(X_test)
metrics.accuracy_score(y_test,y_predict)
```
